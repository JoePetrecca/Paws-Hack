```python
import pandas as pd

# Read the Excel sheet
df = pd.read_excel('liberals.xlsx')

# Extract the second column and remove duplicates
second_column = df.iloc[:, 0].drop_duplicates()

# Create a new DataFrame with only the second column
new_df = pd.DataFrame({'Column 2': second_column})

# Convert to CSV and save to file
new_df.to_csv('liberals.csv', index=False)

# Read the CSV file
with open('liberals.csv') as file:
    # Loop through each line and print it
    for line in file:
        print(line.strip(), '- RACE')
```

```python
import pandas as pd
import re

# Read the Excel sheet into a DataFrame
df = pd.read_excel('liberals.xlsx')
df
looper = []
# Loop over the rows of the DataFrame
for index, row in df.iterrows():
    # Use a regular expression to find course IDs and names in each row
    matches = re.findall(r'[A-Z]{3}\s\d{3}', row.to_string())
    # Print the matches
    for match in matches:
        looper.append(match)
looper
```
